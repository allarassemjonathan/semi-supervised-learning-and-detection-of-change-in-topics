```python
from re import match

topic_txt = open('topic_modeling.txt','r')
topics = {}
cur_topic = None
for line in topic_txt.readlines():
    m = match(' Topic # (\d+) = (\w+)', line)
    if m:
        cur_topic = (m.group(2), int(m.group(1)))
        topics[cur_topic]  = {}
    elif cur_topic is not None:
        m = match("\(\'(\w+)\', (\d+)\)", line)
        if m:
            topics[cur_topic][m.group(1)] = int(m.group(2))
topics
```

```python
topic_unique_words = {}
for id, topic in topics.items():
    topic_unique_words[id] = []
    for word in topic:
        in_others = False
        for id2, topic2 in topics.items():
            if word in topic2 and id != id2:
                in_others = True
                break
        if not in_others:
            topic_unique_words[id].append(word)
# topic_unique_words
all_unique_words = []
for v in topic_unique_words.values():
    for word in v:
        all_unique_words.append(word)
all_unique_words
```

```python
topic_word_probs = {}
goodness_score = {}
topic_score = {}
for id, topic in topics.items():
    topic_score[id] = {}
    topic_count = 0
    for word, count in topic.items():
        topic_count += count
        # if word in all_unique_words:
        #     continue
        tot_count = count
        for id2, topic2 in topics.items():
            if word in topic2 and id != id2:
                tot_count += topic2[word]
        if word not in topic_word_probs:
            topic_word_probs[word] = {}
            goodness_score[word] = {}
        topic_word_probs[word][id] = count / tot_count
        goodness_score[word][id] = count / tot_count * count
        topic_score[id][word] = count / tot_count * count
    # topic_score[id] = {word: count / topic_count for word, count in topic_score[id].items()}
goodness_score
topic_score = {id: {word: score for word, score in sorted(topic.items(), key = lambda x: -x[1])[:5]} for id,  topic in topic_score.items()}
topic_score
```

```python
unique_topic_score_words = {}
for id, topic in topic_score.items():
    unique_topic_score_words[id] = {}
    for word in topic:
        is_unique = True
        for id2, topic2 in topic_score.items():
            if id != id2 and word in topic2: 
                is_unique = False
                break
        if is_unique:
            unique_topic_score_words[id][word] = topic[word]
unique_topic_score_words
```

```python
from itertools import permutations

queries = {}
for id, topic in unique_topic_score_words.items():
    queries[id] = '(' + ') OR ('.join(' '.join(x) for x in permutations(topic.keys(), 2)) + ')'
queries
```
